# Setup

```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
X = pd.read_csv("./data/trnX-3d.csv", sep="\t", header=None)
y = pd.read_csv("./data/trnY-3d.csv", sep="\t", header=None)
dev_X = pd.read_csv("./data/devX-3d.csv", sep="\t", header=None)
dev_y = pd.read_csv("./data/devY-3d.csv", sep="\t", header=None)
```

```python
print(X.shape)
print(y.shape)
print(dev_X.shape)
print(dev_y.shape)
```

# Question 2 Part A

```python
model = MLPClassifier(random_state=42)
model.fit(X, np.array(y).ravel())
```

```python
train_predictions = model.predict(X)
train_accuracy = accuracy_score(y, train_predictions)
train_accuracy
```

```python
test_predictions = model.predict(dev_X)
test_accuracy = accuracy_score(dev_y, test_predictions)
test_accuracy
```

```python
print(f"Accuracy on the Training Data: {train_accuracy} and the Accuracy on the Development Data: {test_accuracy}")
```

# Question 2 Part B

```python
hyper_parameters = {
    'max_iter': 1000,                   
    'activation': 'relu',               
    'solver': 'sgd',                    
    'learning_rate_init': 0.001,        
    'batch_size':32,                    
    'hidden_layer_sizes': (100, 100),
    'random_state': 42             
}
```

```python
mlp_tuned = MLPClassifier(**hyper_parameters)
mlp_tuned.fit(X, np.array(y).ravel())
```

```python
predictions_train = mlp_tuned.predict(X)
train_accuracy_tuned = accuracy_score(y, predictions_train)
train_accuracy_tuned
```

```python
predictions_dev = mlp_tuned.predict(dev_X)
dev_accuracy_tuned = accuracy_score(dev_y, predictions_dev)
dev_accuracy_tuned
```

```python
print(f"Accuracy on the Training Data: {train_accuracy_tuned} and the Accuracy on the Development Data: {dev_accuracy_tuned}")
```
